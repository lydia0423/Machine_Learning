```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## The total number of movies or shows for each rating?

### Pie Chart

```python
netflix = pd.read_csv('netflix_titles.csv')

plt.style.use('ggplot')

tv_ma = netflix.loc[netflix['rating'] == 'TV-MA'].count()[0]
tv_14 = netflix.loc[netflix['rating'] == 'TV-14'].count()[0]
tv_pg = netflix.loc[netflix['rating'] == 'TV-PG'].count()[0]
r = netflix.loc[netflix['rating'] == 'R'].count()[0]
pg_13 = netflix.loc[netflix['rating'] == 'PG-13'].count()[0]

labels= ['TV-MA', 'TV-14', 'TV-PG', 'R', 'PG-13']
explode = [0, 0, 0, 0.4, 0.4]

#labeldistance = None to remove labels from the pie chart and only show in the legend
plt.pie([tv_ma, tv_14, tv_pg, r, pg_13], labels=labels, autopct='%.2f%%', pctdistance=0.8, labeldistance=None)

plt.title('The total number of movies or shows for each rating in Netflix')

plt.legend(title="Ratings", loc="center right", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

### Bar Chart

```python
netflix = pd.read_csv('netflix_titles.csv')

labels= ['TV-MA', 'TV-14', 'TV-PG', 'R', 'PG-13']

tv_ma = netflix.loc[netflix['rating'] == 'TV-MA'].count()[0]
tv_14 = netflix.loc[netflix['rating'] == 'TV-14'].count()[0]
tv_pg = netflix.loc[netflix['rating'] == 'TV-PG'].count()[0]
r = netflix.loc[netflix['rating'] == 'R'].count()[0]
pg_13 = netflix.loc[netflix['rating'] == 'PG-13'].count()[0]
values = [tv_ma, tv_14, tv_pg, r, pg_13]

bars = plt.bar(labels, values, color='green')

patterns = ['/', '-', '.', '*', 'o']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
      
plt.title('The total number of movies or shows for each rating in Netflix')
plt.xlabel('Ratings')
plt.ylabel('Number of movies')

plt.show()

```

## How many movies or shows have multiple seasons?

```python
netflix = pd.read_csv('netflix_titles.csv')

labels = ['1 Season', '2 Seasons', '3 Seasons', '4 Seasons']

a = netflix.loc[netflix['duration'] == '1 Season'].count()[0]
b = netflix.loc[netflix['duration'] == '2 Seasons'].count()[0]
c = netflix.loc[netflix['duration'] == '3 Seasons'].count()[0]
d = netflix.loc[netflix['duration'] == '4 Seasons'].count()[0]
values = [a,b,c,d]

plt.bar(labels, values, color='blue')

plt.title('The Total Number of Movies or Shows Have Multiple Seasons')
plt.xlabel('Seasons')
plt.ylabel('Numbers of movies')

plt.show()
```

## The numbers of movies produced by certain country

```python
netflix = pd.read_csv('netflix_titles.csv')

country_list = ['United States', 'India', 'United Kingdom', 'Japan', 'South Korea']

US = netflix.loc[netflix['country'] == 'United States'].count()[0]
India = netflix.loc[netflix['country'] == 'India'].count()[0]
UK = netflix.loc[netflix['country'] == 'United Kingdom'].count()[0]
JP = netflix.loc[netflix['country'] == 'Japan'].count()[0]
SK = netflix.loc[netflix['country'] == 'South Korea'].count()[0]
values = [US, India, UK, JP, SK]

barh = plt.barh(country_list, values, color='purple')

plt.title('Number of Movies Proeduced by The Given Country in Netflix')
plt.xlabel('Number of movies')
plt.ylabel('Country')

plt.show()
 
```
